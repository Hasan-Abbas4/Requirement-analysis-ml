```python
path1 = "/content/drive/MyDrive/Requirement Analysis/archive (3)/nfr.txt"
path2 = "/content/drive/MyDrive/Requirement Analysis/archive (4)/software_requirements_extended.csv"
path3 = "/content/drive/MyDrive/Requirement Analysis/archive (5)/Pure_Annotate_Dataset.csv"
```

```python
import pandas as pd

data = []

with open(path1, "r", encoding="utf-8") as file:
    for line in file:
        if ":" in line:
            label, sentence = line.split(":", 1)
            label = label.strip()
            sentence = sentence.strip()
            if label in ["F", "FR"]:
                label = 0
            else:
                label = 1

            data.append([sentence, label])

df1 = pd.DataFrame(data, columns=["sentence", "label"])
df2 = pd.read_csv(path2)
df3 = pd.read_csv(path3,encoding='latin-1')
```

```python
def convert_label(x):
    if x in ["F", "FR"]:
        return 0
    else:
        return 1
df2 = df2[["sentence", "label"]]

df3 = df3[["sentence", "NFR_boolean"]]
df3.rename(columns={"NFR_boolean": "label"}, inplace=True)
```

```python
final_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
import re

final_df.dropna(inplace=True)

final_df["sentence"] = final_df["sentence"].str.lower()

final_df["sentence"] = final_df["sentence"].apply(
    lambda x: re.sub(r"[^\w\s]", "", x)
)

final_df.drop_duplicates(inplace=True)
```

```python
save_path = "/content/drive/MyDrive/Requirement Analysis/cleaned_dataset.csv"
final_df.to_csv(save_path, index=False)

print("Saved successfully!")
print("Total rows:", len(final_df))
```
